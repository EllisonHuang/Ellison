A simple linear regression model is the linear relationship between X(the predictor variable/independent variable) and Y(outcome variables/dependent variable) using beta1(the slope/rate of change or the expected change for the next unit of x), beta0(intercept coefficients/telling the start of Y when X is at zero) and sigma(error term/All the y values that are not on the line). 

To form a sample we place all the value of X inside the formula Y = β0 + β1X + ϵ to find all the Y values and placing it on a graph with a line defineded by Y = β0 + β1X + ϵ.

````python
import plotly.express as px
import numpy as np
import pandas as pd

# Define model parameters
n = 50              # Number of data points
x_start = 0         # Start value for X
x_end = 10          # End value for X
beta0 = 5          # Intercept (beta0)
beta1 = 2.5        # Slope (beta1)
sigma = 3           # Standard deviation of the error term

# Generate predictor values (X)
np.random.seed(0)  # For reproducibility
X = np.linspace(x_start, x_end, n)

# Generate random errors with a normal distribution
errors = np.random.normal(0, sigma, n)

# Calculate Y values based on the theoretical model
Y = beta0 + beta1 * X + errors

# Create a DataFrame to hold the data
df = pd.DataFrame({'x': X, 'Y': Y, 'Data': 'Observed'})

# Scatter plot with a trendline using plotly.express
fig = px.scatter(df, x='x', y='Y', color='Data', trendline='ols', title='Y vs. x with Trendline')
fig.show(renderer="png")
````

````python
import statsmodels.formula.api as smf  # This allows us to write a formula where the dependent 
                                       # variable and independent variables are specified.
import plotly.express as px  # this is a ploting library


# First indicate that the model will look at df with the dependent variable Y and the independent variable X. 
# Using the data given the slope is stored as fitted_model.
model_data_specification = smf.ols("Y~x", data=df) 
fitted_model = model_data_specification.fit() 

# what do each of the following provide?
fitted_model.summary()  # This provides the Model Information, R-squared and Adjusted R-squared, 
                        # F-statistic and Prob (F-statistic), Coefficient Table and Diagnostic Information of fitted model.
fitted_model.summary().tables[1]  # This is showing summary of  Coefficient Table including Variable Names, Coefficient, 
                                  # Standard Error, p-value and Confidence Interval.
fitted_model.params  # It uses the independent Variables X to estimate dependent Variables Y using Intercept and slope.
fitted_model.params.values  # Uses independent Variables X and  dependent Variables Y to estimate Intercept and slope.
fitted_model.rsquared  # This returns the R-squared value of the regression model, that is telling how high i can 
                       # how much I can trust the model to explain the variability in Y.

# Creates a column called Data so the data point is used alone. 
# Adding a trend line to show the linear relationship between X and Y
df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 11, explaining how they're similar and different.<br>
    
<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_
>
> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_
> 
>
> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_

````python
import numpy as np
import pandas as pd

# Assuming `old_faithful` has columns `duration` and `kind` (with 'short' and 'long' categories)

# Step 1: Calculate the observed difference in means
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Step 2: Set up permutation test parameters
num_permutations = 10000  # Number of permutations
permuted_diffs = []

# Step 3: Perform the permutation test
for _ in range(num_permutations):
    # Shuffle the 'kind' labels and assign them to a new column
    old_faithful_shuffled = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)
    
    # Calculate mean difference for this shuffle
    permuted_diff = old_faithful_shuffled.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(permuted_diff)

# Step 4: Convert list to numpy array for easier manipulation
permuted_diffs = np.array(permuted_diffs)

# Step 5: Calculate p-value: Proportion of permuted differences that are as extreme as the observed difference
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))

# Display the results
print(f"Observed difference in means: {observed_diff}")
print(f"P-value from permutation test: {p_value}")
````

````python
import numpy as np
import pandas as pd

# Assuming `old_faithful` has columns `duration` and `kind` (with 'short' and 'long' categories)

# Set up bootstrap parameters
num_bootstraps = 10000  # Number of bootstrap samples
bootstrapped_mean_diffs = []

# Perform the two-sample bootstrap procedure
for _ in range(num_bootstraps):
    # Bootstrap sample within each group
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    
    # Calculate the mean difference for this bootstrap sample
    bootstrapped_diff = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_mean_diffs.append(bootstrapped_diff)

# Convert list to numpy array for easier manipulation
bootstrapped_mean_diffs = np.array(bootstrapped_mean_diffs)

# Calculate the 95% confidence interval
confidence_interval = np.quantile(bootstrapped_mean_diffs, [0.025, 0.975])

# Display the results
print(f"95% Bootstrap Confidence Interval for the difference in means: {confidence_interval}")
````

For the shuffling we assume there is no difference between the groups, we use any natural association between the label and the duration by randomly shuffling the labels. > - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
