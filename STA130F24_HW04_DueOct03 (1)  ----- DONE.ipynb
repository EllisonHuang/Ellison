````python
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([5, 15, 25, 35, 45, 55, 65, 75, 85, 95])

# Number of times it would repeat the data
n_bootstrap_samples = 1000

# Bootstrap the data
bootstrap_means = []
for _ in range(n_bootstrap_samples):
    #bootstrapping the data
    sample = np.random.choice(data, size=len(data), replace=True)
    #finding the mean of the sample
    sample_mean = np.mean(sample)
    #place the mean inside the list
    bootstrap_means.append(sample_mean)

# Calculate the 95% confidence interval
lower_bound = np.quantile(bootstrap_means, 0.025)
upper_bound = np.quantile(bootstrap_means, 0.975)

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, color='skyblue', alpha=0.7, label='Bootstrapped Means')
plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound:.2f})')
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound ({upper_bound:.2f})')
plt.title('Distribution of Bootstrapped Means and 95% Confidence Interval')
plt.xlabel('Mean Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([5, 15, 25, 35, 45, 55, 65, 75, 85, 95])

# Number of bootstrap samples
n_bootstrap_samples = 1000

# Bootstrap the data
bootstrap_medians = []
for _ in range(n_bootstrap_samples):
    sample = np.random.choice(data, size=len(data), replace=True)
    #finding the meian of the sample
    sample_median = np.median(sample)
    bootstrap_medians.append(sample_median)

# Calculate the 95% confidence interval
lower_bound = np.quantile(bootstrap_medians, 0.025)
upper_bound = np.quantile(bootstrap_medians, 0.975)

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_medians, bins=30, color='skyblue', alpha=0.7, label='Bootstrapped Medians')
plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound:.2f})')
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound ({upper_bound:.2f})')
plt.title('Distribution of Bootstrapped Medians and 95% Confidence Interval')
plt.xlabel('Median Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([5, 15, 25, 35, 45, 55, 65, 75, 85, 95])

# Number of bootstrap samples
n_bootstrap_samples = 1000

# Bootstrap the data to calculate standard deviations
bootstrap_std_devs = []
for _ in range(n_bootstrap_samples):
    sample = np.random.choice(data, size=len(data), replace=True)
    sample_std_dev = np.std(sample, ddof=1)  # Using ddof=1 for sample standard deviation to make sure is the standard deviation of the entire data
    bootstrap_std_devs.append(sample_std_dev)

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrap_std_devs, 2.5)
upper_bound = np.percentile(bootstrap_std_devs, 97.5)

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_std_devs, bins=30, color='skyblue', alpha=0.7, label='Bootstrapped Standard Deviations')
plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound:.2f})')
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound ({upper_bound:.2f})')
plt.title('Distribution of Bootstrapped Standard Deviations and 95% Confidence Interval')
plt.xlabel('Standard Deviation Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Outputting the results
print(f"95% Confidence Interval for the standard deviation: [{lower_bound:.2f}, {upper_bound:.2f}]")
````

````python
np.random.seed(42)

# Initial Health Scores (normally distributed)
initial_scores = np.random.normal(loc=80, scale=5, size=100)  # Mean 80, std dev 5

# Final Health Scores (close to initial scores, small random variation)
final_scores = initial_scores + np.random.normal(loc=0, scale=2, size=100)  # Small changes, mean 0, std dev 2

# Bootstrapping function with both mean and SD
def bootstrap_with_mean_sd(data, n_bootstraps=1000):
    boot_means = []
    np.random.seed(42)  # For reproducibility
    for _ in range(n_bootstraps):
        sample = np.random.choice(data, size=len(data), replace=True)
        boot_means.append(np.mean(sample))
    mean = np.mean(boot_means)
    std_dev = np.std(boot_means, ddof=1)  # Standard Deviation of the bootstrapped means
    return boot_means, mean, std_dev

# Perform bootstrapping for both initial and final scores
bootstrapped_initial, mean_initial_bootstrap, std_initial_bootstrap = bootstrap_with_mean_sd(initial_scores)
bootstrapped_final, mean_final_bootstrap, std_final_bootstrap = bootstrap_with_mean_sd(final_scores)

# Calculate original means for reference
mean_initial_original = np.mean(initial_scores)
mean_final_original = np.mean(final_scores)

# Plotting the results
plt.figure(figsize=(14, 12))

# Plot the original Initial Health Score distribution as a histogram
plt.subplot(2, 2, 1)
plt.hist(initial_scores, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(x=mean_initial_original, color='k', linestyle='-', label=f'Original Mean: {mean_initial_original:.2f}')
plt.title('Original Initial Health Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()

# Plot the original Final Health Score distribution as a histogram
plt.subplot(2, 2, 2)
plt.hist(final_scores, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.axvline(x=mean_final_original, color='k', linestyle='-', label=f'Original Mean: {mean_final_original:.2f}')
plt.title('Original Final Health Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()

# Plot the bootstrapped Initial Health Score means distribution as a histogram (±2*SD only)
plt.subplot(2, 2, 3)
plt.hist(bootstrapped_initial, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(x=mean_initial_bootstrap, color='k', linestyle='-', label=f'Bootstrapped Mean: {mean_initial_bootstrap:.2f}')
plt.axvline(x=mean_initial_bootstrap - 2*std_initial_bootstrap, color='b', linestyle='--', label='Mean ± 2*SD')
plt.axvline(x=mean_initial_bootstrap + 2*std_initial_bootstrap, color='b', linestyle='--')
plt.title('Bootstrapped Initial Health Score Means')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.legend()

# Plot the bootstrapped Final Health Score means distribution as a histogram (±2*SD only)
plt.subplot(2, 2, 4)
plt.hist(bootstrapped_final, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.axvline(x=mean_final_bootstrap, color='k', linestyle='-', label=f'Bootstrapped Mean: {mean_final_bootstrap:.2f}')
plt.axvline(x=mean_final_bootstrap - 2*std_final_bootstrap, color='b', linestyle='--', label='Mean ± 2*SD')
plt.axvline(x=mean_final_bootstrap + 2*std_final_bootstrap, color='b', linestyle='--')
plt.title('Bootstrapped Final Health Score Means')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()
````
