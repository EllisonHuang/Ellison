1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

# summary:
Summary of Metrics and When to Use Them
Accuracy

What it measures: Overall correctness (TP + TN / Total).
Use Case: Balanced datasets where all errors have similar costs.
Example: Diagnosing minor illnesses.
Limitation: Misleading for imbalanced data or unequal error costs.
Sensitivity (Recall)

What it measures: Proportion of actual positives correctly identified (TP / TP + FN).
Use Case: Critical to minimize false negatives.
Example: Cancer screening.
Limitation: Can increase false positives, lowering precision.
Specificity

What it measures: Proportion of actual negatives correctly identified (TN / TN + FP).
Use Case: Critical to minimize false positives.
Example: Fraud detection.
Limitation: Can increase false negatives, lowering sensitivity.
Precision

What it measures: Proportion of positive predictions that are correct (TP / TP + FP).
Use Case: Critical to minimize false positives.
Example: Targeted advertising.
Limitation: Can miss positives, lowering recall.
Key Point:
Choose metrics based on the importance of false positives vs. false negatives and dataset balance.

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd

X = ab_reduced_noNaN[['List Price']]
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']

# Train a classification tree with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)  # Import fixed
clf.fit(X, y)

# Visualize the tree
fig, ax = plt.subplots(figsize=(10, 7))  # Create a figure and axes
plot_tree(
    clf,
    feature_names=['List Price'],
    class_names=['Paperback', 'Hardcover'],
    filled=True,
    rounded=True,
    ax=ax  # Pass the axes to plot_tree
)

# Save the figure as a PNG file
fig.savefig("tree_visualization.png", dpi=300)
````

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset: Adjust this to your actual dataset
# Define X and y
X_clf = ab_reduced_noNaN[['List Price']]  # Features for clf
X_clf2 = ab_reduced_noNaN[['NumPages', 'Thick', 'List Price']]  # Features for clf2
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']

# Split data into training and testing sets
X_train_clf, x_test_clf, y_train, y_test = train_test_split(X_clf, y, test_size=0.3, random_state=42)
X_train_clf2, x_test_clf2, _, _ = train_test_split(X_clf2, y, test_size=0.3, random_state=42)

# Train the classifiers
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train_clf, y_train)

clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X_train_clf2, y_train)

# Predictions for clf and clf2
y_pred_clf = clf.predict(x_test_clf)
y_pred_clf2 = clf2.predict(x_test_clf2)

# Confusion Matrices
conf_mat_clf = confusion_matrix(y_test, y_pred_clf)
conf_mat_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Plot the confusion matrix for clf
fig1, ax1 = plt.subplots(figsize=(8, 6))
disp_clf = ConfusionMatrixDisplay(confusion_matrix=conf_mat_clf, display_labels=['Paperback', 'Hardcover'])
disp_clf.plot(cmap=plt.cm.Blues, ax=ax1)
ax1.set_title("Confusion Matrix for clf")
fig1.savefig("confusion_matrix_clf.png", format="png", dpi=300)  # Save as PNG

# Plot the confusion matrix for clf2
fig2, ax2 = plt.subplots(figsize=(8, 6))
disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=conf_mat_clf2, display_labels=['Paperback', 'Hardcover'])
disp_clf2.plot(cmap=plt.cm.Greens, ax=ax2)
ax2.set_title("Confusion Matrix for clf2")
fig2.savefig("confusion_matrix_clf2.png", format="png", dpi=300)  # Save as PNG

# Display Metrics for clf
tn, fp, fn, tp = conf_mat_clf.ravel()
sensitivity_clf = tp / (tp + fn)
specificity_clf = tn / (tn + fp)
accuracy_clf = accuracy_score(y_test, y_pred_clf)
print(f"clf Metrics: Sensitivity: {sensitivity_clf:.2f}, Specificity: {specificity_clf:.2f}, Accuracy: {accuracy_clf:.2f}")

# Display Metrics for clf2
tn2, fp2, fn2, tp2 = conf_mat_clf2.ravel()
sensitivity_clf2 = tp2 / (tp2 + fn2)
specificity_clf2 = tn2 / (tn2 + fp2)
accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)
print(f"clf2 Metrics: Sensitivity: {sensitivity_clf2:.2f}, Specificity: {specificity_clf2:.2f}, Accuracy: {accuracy_clf2:.2f}")
````

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).
